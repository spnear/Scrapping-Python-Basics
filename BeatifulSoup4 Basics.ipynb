### BeatifulSoup 

Librer√≠a utilizada para manipular el documento HTML, convierte el documento en una estructura para manipular con queries

```python
import bs4
import requests
```

```python
url = 'https://www.platzi.com'
response = requests.get(url)
response.encoding ='utf-8'
```

```python
soup = bs4.BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.title.text)
```

```python
print(soup.select('meta[name=description]'))
```

```python
print(soup.select('meta[name=description]')[0]['content'])
```

```python
courses_link = soup.select('.SchoolsList-school')
courses = [course['href'] for course in courses_link]

for course in courses:
    print(course)
```

#### Reto personal:

Hago un scrapping sencillo con lo aprendido con la p√°gina web de un peri√≥dico local (La naci√≥n - Neiva)
Extraigo la lista de noticias y seguido extraigo datos de un art√≠culo

```python
url = "https://www.lanacion.com.co/"

response = requests.get(url)

soup = bs4.BeautifulSoup(response.text, 'html.parser')
news = soup.select('.penci_media_object a')

for item in news:
    print(f"Noticia: {item.text} | Link: {url}{item['href']}")
    print()
```

```python
url = "https://www.lanacion.com.co/huila-registro-17-muertes-por-covid-19-2/"

response = requests.get(url)

soup = bs4.BeautifulSoup(response.text, 'html.parser')
paragraphs = soup.select('.penci-container p')

#print(soup)
for idx,item in enumerate(paragraphs):
    print(f"P√°rrafo No{idx}, texto: \n{item.text}\n")
    print(item)
```

```python
#Extraer t√≠tulo
title = soup.select('h1.penci-entry-title')
print(title[0].text)
```
